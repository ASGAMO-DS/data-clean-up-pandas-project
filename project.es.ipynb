#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
# Código

from statistics import mean

CalificacionesEstudiantes = [8.5,9.2,7.8,8.9,10]
promedio = mean(CalificacionesEstudiantes)

print(f'el promedio de calificación de los estudiantes es: {promedio}')
```

- Encuentra la calificación más alta y la más baja.

```python
# Código
CalificaciónMasAlta = max(CalificacionesEstudiantes)
CalificaciónMásBaja = min(CalificacionesEstudiantes)

print(f'La Calificación más Alta es: {CalificaciónMasAlta}')
print(f'La Calificación más Baja es: {CalificaciónMásBaja}')
```

#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
# TODO

casa_mas_cara = ds.loc[ds['price'].idxmax()]
print(f"La casa más cara es la que se encuentra en: {casa_mas_cara['address']}")
print(f"Precio: ${casa_mas_cara['price']:,.2f}")
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
# TODO

# Filtrar por la población "Arroyomolinos (Madrid)"
ds_arroyomolinos = ds[ds['level5'] == "Arroyomolinos (Madrid)"]

# Calcular la media de precios
media_precio = ds_arroyomolinos['price'].mean()

# Imprimir el resultado
print(f"La media de precios en Arroyomolinos (Madrid) es: {media_precio:.2f} USD")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt

# Trazar el histograma de precios

ds_arroyomolinos = ds[ds['level5'] == "Arroyomolinos (Madrid)"]

# Graficar el histograma de precios
plt.figure(figsize=(10, 5))
plt.hist(ds_arroyomolinos['price'], bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Precio (USD)')
plt.ylabel('Frecuencia')
plt.title('Histograma de precios en Arroyomolinos (Madrid)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Se evidencia que los precios en Arroyomolinos tiene una distribución casi normal
